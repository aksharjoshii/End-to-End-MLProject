```python
import pandas as pd 
```

```python
df1 = pd.read_csv('2017_Yellow_Taxi_Trip_Data.csv')
df2 = pd.read_csv('nyc_preds_means.csv')
df = df1.merge(df2, left_index=True, right_index=True)
df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
pd.set_option('display.max_columns', None)
```

```python
df1 = df[df['payment_type']==1]
df1.reset_index()
df1['tip_percent'] = round(df1['tip_amount'] / (df1['total_amount'] - df1['tip_amount']), 3)
df1['generous'] = df1['tip_percent']
df1['generous'] = (df1['generous'] >= 0.2)
df1.generous =df1.generous.astype(int)
df1.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_df , test_df  = train_test_split(df1, test_size=0.05, random_state=0)
```

```python
train_df.shape
```

```python
train_df.to_csv('nyx_taxi_data.csv', index_label=False)
test_df.to_csv('test_data.csv', index_label=False)
```
